```python
x = []
y= []
with open("p4_data_13.csv") as file:
    print(file.readline())
    for line in file:
        line = line.strip().split(",")
        for el in line:
            
            el.replace("$", '')
        
        x.append(float(line[4].replace("$", "")))
        y.append(float(line[6].replace("$", "")))

print(x)
print(y)
```

```python
print(len(x))
```

```python

x1 = range(0,1000)
plt.figure(figsize=(15,10))
plt.bar(x1,x)
plt.title("Гистограмма зарплат")
plt.ylabel("Зарплата")
```

```python
print(x)
print(y, "\n")
array1 = np.array(x)
array2 = np.array(y)

temp1 = array1.argsort()
temp2 = array2.argsort()
arr1 = []
arr2 = []
for i in temp1:
    arr1.append(array1[i])
for j in temp2:
    arr2.append(array2[j])
print(arr1)
print(arr2)
```

```python

plt.figure(figsize=(10,10))
plt.plot(arr2,arr1)
plt.title("График зависимости налога от зарплаты")
plt.xlabel("Налог")
plt.ylabel("Зарплата")
```

1. NumPy предоставляет широкий спектр математических функций, которые делают сложные численные операции простыми. В списках Python нет таких специфических численных функций, и для выполнения подобных операций требуется больше ручного труда.
Список Python может содержать разные типы данных, а массив NumPy — только одного типа. 



```python
p = [1,2,3,45,6,7,8,765,4,6,54,35,43,6,76,543,23,45]
p = np.array(p)
print(p, type(p), "\n")

p1 = np.arange(0, 10, 1)
print(p1, type(p1), "\n")

print(np.zeros(2, dtype="int32"))
print(np.ones(2, dtype="int32"))


print(p * 2)
print(p + p, "\n")

print(np.random.rand(4), "\n")
print(np.random.randint(low= 0, high=100, size=(10)), '\n')
print(np.random.normal(loc = 50, scale= 50, size = 100))

```
