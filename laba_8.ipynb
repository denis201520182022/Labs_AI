```python
import requests
from matplotlib import pyplot as plt
import numpy as np
```

```python


url = "https://raw.githubusercontent.com/enikolaev/IT-and-Programming/refs/heads/main/data/pract8/pr8_13.txt"

response = requests.get(url)

with open('pr8_13.txt', 'wb') as file:
    file.write(response.content)
```

```python
f  = open("pr8_13.txt")
s = f.readlines()
x =[]
y =[]
for line in s:
    # print(line, type(line))
    d = line.strip().split(' ')
    x.append(float(d[0]))
    y.append(float(d[1]))

print(x, "\n", y)
```

```python
plt.figure()
plt.scatter(x,y, c="g", marker="D", s=5, alpha = 0.5 )
```

```python
X = np.array(x)
Y = np.array(y)
```

```python
k, b = 0.0, 1.0
xmin, xmax = min(X), max(X)

alpha = 0.1
N = 20
for i in range(N):
    y_pred = k*X + b
    
    MSE = np.sum((Y - y_pred)**2) / Y.shape[0]
    plt.figure(figsize=(6,4))
    plt.scatter(X, y, color="red", s = 2)
    plt.plot([xmin, xmax], [k*xmin + b, k*xmax + b])
    plt.title(f'№{i}, MSE = {MSE}, k = {k:10f}, b = {b:10f}' )
    plt.show()
    gk = - ( 2 * np.dot(X.T, (Y - y_pred))) / Y.shape[0]
    gb = - 2 * np.sum(Y - y_pred) / Y.shape[0]
    
    k = float(k - alpha * gk)
    b = b - alpha * gb
```

```python
X1 = np.zeros(15)
for i in range(3, 15):
    X1[i - 3] = i
    print(X1, type(x))
X1 = np.array(X1)
y_pred = k*X1 + b

plt.figure(figsize=(10,15))
plt.plot(X1,y_pred)
plt.scatter(X, Y, c="g", marker="o", s=5, alpha = 0.5 )


```
